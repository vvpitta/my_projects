```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2013, 1, 27)

f = web.DataReader("F", 'google', start, end)
print(f)
```

```python
closePrice = quote.loc['Close']
print(closePrice)
```

```python
openPriceData = f['Open'].iloc[:]
closePriceData = f['Close'].iloc[:]

openPrice = list(openPriceData)
closePrice = list(closePriceData)

dates = []
for x in range(len(f)):
    newdate = str(f.index[x])
    newdate = newdate[0:10]
    dates.append(newdate)
    
f['Date'] = dates
```

```python
trace0 = go.Scatter(
    x = dates,
    y = openPrice,
    name = 'Open Price',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 4)
)

trace1 = go.Scatter(
    x = dates,
    y = closePrice,
    name = 'Close Price',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 4)
)

data = [trace0, trace1]

layout = dict(title = 'Ford Stock Price',
             xaxis = dict(title = 'Date'),
             yaxis = dict(title = 'Price'),
             )
 
fig = dict(data = data, layout = layout)
py.iplot(fig, filename = 'styled-line')
```

```python
edgarRequest = requests.get("http://edgaronline.api.mashery.com/v2/corefinancials/ann.json?primarysymbols=msft&appkey=exnfjwc3qfd6ba6m9hbguqys")
edgarRequest.text

edgarData = json.loads(edgarRequest.text)
print(edgarData)
```

```python
nest = edgarData['result']['rows']

# table = PrettyTable(['Field', 'Value'])
# table.add_row([field['field'], field['value']])
# print(table)

fiscalYears = []
currentAssets = []
currentLiab = []
inventories = []
totalRevenue = []
cogs = []
receivables = []

for item in nest:
    for field in item['values']:
        if field['field'] == 'fiscalyear':
            fiscalYears.append(field['value'])
        if field['field'] == 'totalcurrentassets':
            currentAssets.append(field['value'])
        if field['field'] == 'totalcurrentliabilities':
            currentLiab.append(field['value'])
        if field['field'] == 'inventoriesnet':
            inventories.append(field['value'])
        if field['field'] == 'totalrevenue':
            totalRevenue.append(field['value'])
        if field['field'] == 'costofrevenue':
            cogs.append(field['value'])
        if field['field'] == 'totalreceivablesnet':
            receivables.append(field['value'])

print('Year:', fiscalYears)
print('Total Current Assets:', currentAssets)
print('Total Current Liabilities:', currentLiab)
print('Inventories:', inventories)
print('Total Revenue:', totalRevenue)
print('Cost of Goods Sold:', cogs)
print('Total Receivables:', receivables)
```

```python
currentRatios = []
quickRatios = []
inventoryTurnover = []
dso = []

# Liquidity Ratios
for x in range(0,4):
    currentRatios.append(currentAssets[x]/currentLiab[x])
    quickRatios.append((currentAssets[x] - inventories[x])/currentLiab[x])
    
print('Current Ratio:', currentRatios)
print('Quick Ratio:', quickRatios)
    
# Asset Management Ratios
for x in range(0,4):
    inventoryTurnover.append(cogs[x]/inventories[x])
    dso.append(receivables[x]/(totalRevenue[x]/365))
    
print('Inventory Turnover Ratio:', inventoryTurnover)
print('DSO:', dso)
```
